Q5. Problem: A student has name and attendance percentage.

Write a class to store and display student details.

- define class Student

- define __init__ method

- define display method

create object and display details


```python
class Student:
    def __init__(self, name, attendance):
        self.name = name
        self.attendance = attendance

    def display(self):
        print("Student Name:", self.name)
        print("Attendance:", self.attendance, "%")

s1 = Student("Anita", 92)
s1.display()
```
