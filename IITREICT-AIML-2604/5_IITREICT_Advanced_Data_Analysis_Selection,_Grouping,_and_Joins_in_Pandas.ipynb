## 1. Why Do We Need Grouping?

Consider the following employee dataset:

```python
import pandas as pd

data = {
    "Employee": ["Amit", "Maya", "Kiran", "Sarah", "Alex"],
    "Department": ["HR", "IT", "IT", "HR", "Finance"],
    "Salary": [45000, 70000, 72000, 50000, 65000]
}

df = pd.DataFrame(data)

print(df)
```

| -------- | ---------------- |

| mean()   | Average          |
| count()  | Number of values |

# Topic 4: Combining DataFrames and Structural Joins

- In **real-world data analysis**, information is often **stored across multiple tables** rather than a s**ingle DataFrame**.

- **For example:** **Employee** details may be stored in **one** table. **Department** information may be stored in **another** table. **Salary** records may be stored in a **third** table.

- To perform meaningful analysis, **these tables must be combined**. Pandas provides powerful functions such as **merge(), concat(), and join()** to combine DataFrames efficiently.
- This **concept** is **similar to SQL joins** and is widely used in data engineering, analytics, and machine learning pipelines.

## 1. Why Do We Need Combining DataFrames?

Consider the following situation.



```python
# Employee Table
import pandas as pd

employees = pd.DataFrame({
    "Emp_ID": [101, 102, 103],
    "Name": ["Amit", "Maya", "Kiran"]
})

print(employees)
```

```python
# Department Table
departments = pd.DataFrame({
    "Emp_ID": [101, 102, 103],
    "Department": ["HR", "IT", "Finance"]
})

print(departments)
```

**Question:** How do we get Employee Name and Department together?

**Answer:** merge()

## 2. Understanding merge()

The **merge()** function **combines DataFrames** using a **common column**.

**Syntax**

pd.merge(
    left_df,
    right_df,
    on="common_column"
)

```python
# Syntax

pd.merge(
    left_df,
    right_df,
    on="common_column"
)
```

## 3. Inner Join - Matching rows only

```python
merged_df = pd.merge(
    employees,
    departments,
    on="Emp_ID"
)

print(merged_df)
```

**Visual Understanding**
*Table A*
- 101
- 102
- 103

*Table B*
- 101
- 102
- 104

*Inner Join Result*
- 101
- 102

## 4. Types of Joins

Pandas supports four major joins:

| Join  | Description               |
| ----- | ------------------------- |
| inner | Matching rows only        |
| left  | All rows from left table  |
| right | All rows from right table |
| outer | All rows from both tables |


## 5. Left Join - All rows from left table

Keep **everything** from the **left DataFrame**.

```python
# Example

employees = pd.DataFrame({
    "Emp_ID":[101,102,103],
    "Name":["Amit","Maya","Kiran"]
})

departments = pd.DataFrame({
    "Emp_ID":[101,102],
    "Department":["HR","IT"]
})

pd.merge(
    employees,
    departments,
    on="Emp_ID",
    how="left"
)
```

 **Notice** : 103 was kept because it exists in the left table.

## 6. Right Join - All rows from right table

Keep everything from the right table.

```python
pd.merge(
    employees,
    departments,
    on="Emp_ID",
    how="right"
)
# Only rows present in the right table are guaranteed to appear.
```

## 7. Outer Join - All rows from both tables

Keep everything from both tables.

```python
pd.merge(
    employees,
    departments,
    on="Emp_ID",
    how="outer"
)
# All rows from both tables.
# Missing values become: NaN
```

**Which join keeps all records?**

**Answer**: Outer Join

**Which join keeps only matching records?**

**Answer**: Inner Join

## 8. Understanding concat()

While merge **combines tables** using **common columns**, **merge()** looks for relationships.

**concat()** - simply stacks DataFrames.

```python
# Example

df1 = pd.DataFrame({
    "Name":["Amit","Maya"]
})

df2 = pd.DataFrame({
    "Name":["Kiran","Sarah"]
})

pd.concat([df1, df2])
```

**Reset Index During Conca**t

```python
pd.concat(
    [df1, df2],
    ignore_index=True
)
# clear indexing
```

## 9. Vertical vs Horizontal Concatenation

```python
# Vertical default , Rows increase
pd.concat(
    [df1, df2]
)
```

```python
# Horizontal - Columns increase
pd.concat(
    [df1, df2],
    axis=1
)
```

## Difference Between merge() and concat()

| merge()                 | concat()                |
| ----------------------- | ----------------------- |
| Uses common key         | No common key needed    |
| Similar to SQL Join     | Similar to stacking     |
| Combines related tables | Combines similar tables |

```python
import pandas as pd

import numpy as np

data = {

'employee': ['Amit', 'Maya', 'Alex', 'Kiran', 'Sarah', 'Kumar'],

'department': ['HR', 'IT', 'Finance', 'IT', 'Marketing', 'HR'],

'salary': [45000, 70000, 65000, 72000, 50000, 48000],

'experience': [2, 5, 2, 6, 3, 1]

}

df = pd.DataFrame(data)



# Filter employees with salary > 50000 and experience > 2

filtered_df = df[(df["salary"] > 50000) & (df["experience"] > 2)]



sorted_df = filtered_df.sort_values(by="salary", ascending=False)



# Reorder columns

final_df = sorted_df[["department", "employee", "salary", "experience"]]



# Compute mean salary using NumPy

mean_salary = np.mean(final_df["salary"])



# Display results

print("Filtered and Sorted Report:")

print(final_df)

print("\nMean Salary of Filtered Employees:")

print(mean_salary)
```

<details><summary><b>Ideal Approach Walkthrough</b></summary>

- **Step 1** — Filter with AND. Apply a combined Boolean filter in a single expression. Each sub-condition must be
wrapped in its own parentheses to enforce correct operator precedence. df['salary'] > 50000 produces a Boolean Series; df['experience'] > 2 produces another. The & operator combines them element-wise:
only rows where both are True survive. Amit (salary 45000), Alex (experience exactly 2 — not strictly greater), Sarah (salary exactly 50000 — not strictly greater), and Kumar (salary 48000) are all excluded. Maya (70000, experience 5) and Kiran (72000, experience 6) pass both conditions.

- **Step 2** — Sort descending. Call sort_values(by='salary', ascending=False) on the filtered
DataFrame. Kiran (72000) ranks above Maya (70000), so Kiran's row appears first.
- **Step 3** — Reorder columns. Pass the desired column order as a Python list inside double brackets. This
produces a new DataFrame view with columns in the order department , employee , salary ,

- **Step 4** — Compute mean. Pass the salary column of the result to np.mean() . The arithmetic mean of

```python
import pandas as pd
import numpy as np
data = {
'employee': ['Amit', 'Maya', 'Alex', 'Kiran', 'Sarah', 'Kumar'],
'department': ['HR', 'IT', 'Finance', 'IT', 'Marketing', 'HR'],
'salary': [45000, 70000, 65000, 72000, 50000, 48000],
'experience': [2, 5, 2, 6, 3, 1]
}
df = pd.DataFrame(data)
# Step 1: Filter — salary strictly above 50000 AND experience strictly above 2
filtered_df = df[(df['salary'] > 50000) & (df['experience'] > 2)]
# Step 2: Sort by salary descending
sorted_df = filtered_df.sort_values(by='salary', ascending=False)
# Step 3: Reorder columns
result = sorted_df[['department', 'employee', 'salary', 'experience']]
# Step 4: Print result and mean salary
print(result)
print(np.mean(result['salary']))
```

## Question 2

**Problem Statement:** A **data analyst** at an **online education platform** is building a report to identify **high-performing students** across **departments**. The platform `stores data in three separate tables`: a `student registration table`, a `marks table`, and a `department table`. The analyst must `merge all three tables`, `filter to students` who `scored above 80`, and then compute the `average score per department` using a chained merge workflow.

**Constraints & Requirements:**

- Use pandas only `(import pandas as pd)`; no other libraries are required.
- Merge the three DataFrames using chained `pd.merge()` calls.
- Use an `inner join` when merging the student table with the marks table (only students with scores should appear).
- Use a `left join` when merging the intermediate result with the department table `(preserve all matched students even if a department record is somehow missing)`.
- After merging, `filter the result` to rows where `marks > 80` .
- Compute the `average marks` per dept_name using `.groupby()` .
- Handle the case where the marks and department tables use different key column names from the student table `(marks uses sid , department uses dept_id )`.

```python
# Sample Data & Inputs:

import pandas as pd
# Table 1: Student registration
students = pd.DataFrame({
"student_id": [1, 2, 3, 4, 5],
"name": ["Student A", "Student B", "Student C", "Student D", "Student E"],
"dept_id": [10, 20, 10, 30, 20]
})
# Table 2: Marks (key column is named 'sid', not 'student_id')
marks = pd.DataFrame({
"sid": [1, 2, 3, 5, 6],
"marks": [92, 75, 88, 61, 95]
})
# Table 3: Department (key column is 'dept_id')
departments = pd.DataFrame({
"dept_id": [10, 20, 30],
"dept_name": ["Computer Science", "Mechanical", "Electrical"]
})
```

```python
merged_df = pd.merge(students, marks, left_on="student_id", right_on="sid", how="inner")
print(merged_df)

merged_df = pd.merge(merged_df, departments, on="dept_id", how="left")
print(merged_df)

filtered_df = merged_df[merged_df['marks'] > 80]
print(filtered_df)

average_marks_per_dept = filtered_df.groupby('dept_name')['marks'].mean().reset_index()
print(average_marks_per_dept)
# Display the final result

print(average_marks_per_dept)
```

<details><summary><b>Ideal Approach Walkthrough</b></summary>

1. **Step 1** — Inner join (students + marks with mismatched key names): The student registration table uses
student_id as the key; the marks table uses sid . Because the key columns have different names,
left_on="student_id" and right_on="sid" are required. Using how="inner" keeps only
students who appear in both tables. Students 1, 2, 3, and 5 match. Student 4 has no marks entry (dropped).
Student 6 has marks but no registration record (dropped by the inner join).
2. **Step 2** — Left join (intermediate result + departments): The intermediate result from step 1 contains
dept_id (from students). Merging with departments on dept_id using how="left" resolves
the department name for each matched student. A left join ensures that even if a dept_id is missing from
the departments table, the student row is still retained (with NaN for dept_name ).
3. **Step 3** — Filter marks > 80: Apply a boolean condition to keep only rows where marks > 80 . Students
1 (marks=92) and 3 (marks=88) pass; students 2 (marks=75) and 5 (marks=61) are dropped.
4. **Step 4** — GroupBy and average: Group the filtered DataFrame by dept_name and compute the mean of
the marks column. Both remaining students belong to "Computer Science". Average = (92 + 88) / 2 =
90.0.
<details>

```python
import pandas as pd
# Data setup
students = pd.DataFrame({
"student_id": [1, 2, 3, 4, 5],
"name": ["Student A", "Student B", "Student C", "Student D", "Student E"],
"dept_id": [10, 20, 10, 30, 20]
})
marks = pd.DataFrame({
"sid": [1, 2, 3, 5, 6],
"marks": [92, 75, 88, 61, 95]
})
departments = pd.DataFrame({
"dept_id": [10, 20, 30],
"dept_name": ["Computer Science", "Mechanical", "Electrical"]
})
# Step 1: Merge students and marks — inner join, mismatched key names
step1 = pd.merge(students, marks, left_on="student_id", right_on="sid", how="inner")
# Step 2: Merge with departments — left join on dept_id
step2 = pd.merge(step1, departments, on="dept_id", how="left")
# Step 3: Filter to students scoring above 80
filtered = step2[step2["marks"] > 80]
# Step 4: Average marks per department
avg_by_dept = filtered.groupby("dept_name")["marks"].mean()
print(avg_by_dept)
```
