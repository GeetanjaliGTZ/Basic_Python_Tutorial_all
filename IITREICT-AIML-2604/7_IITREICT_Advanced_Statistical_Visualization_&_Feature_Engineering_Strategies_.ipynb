# Topic 5: Feature Scaling — Min-Max Scaling and Standard Scaling

This is one of the most important topics for assignments, interviews, and machine learning practice because many algorithms perform poorly when features have different scales.

**Topic 5: Feature Scaling — Min-Max Scaling and Standard Scaling**

This is one of the most important topics for machine learning practice because many algorithms perform poorly when features have different scales.

**1. What is Feature Scaling?**

**Definition:** Feature Scaling is the **process of transforming numerical features** so they have comparable ranges or distributions.

**Why?**

Suppose we have:
| Age | Salary |
| --- | ------ |
| 25  | 100000 |
| 30  | 150000 |

Salary values are thousands of times larger than age values.

As a result, many ML algorithms **treat salary as more important than age**.

**2. Why Do We Need Scaling?**

The tutorial states that distance-based algorithms interpret larger values as more important.

**Without scaling:**

Large-valued feature, dominates the model

**With scaling:** All features contribute fairly

**Example**: **Before Scaling**
| Age | Salary  |
| --- | ------- |
| 25  | 100000  |
| 50  | 1500000 |

Huge difference in scales.

**After Scaling**

| Age | Salary |
| --- | ------ |
| 0.2 | 0.3    |
| 0.8 | 1.0    |


**3. Min-Max Scaling**

**Definition**: Min-Max Scaling transforms data into the range: [0,1]

Xscaled ​ =Xmax ​− Xmin/ ​X − Xmin​​

**Important Properties**

- Minimum value becomes: 0
- Maximum value becomes: 1
- All other values lie between: 0 and 1

#Example

Example

Suppose: Feature = [10,20,30,40,50]

Here: **Min** = 10, **Max** = 50

For value 30:

50−10 = 40, 30−10=20
	​
20/40 = 0.5

**Scaled value:**

0.5

```python
# Original Feature
feature = [10, 20, 30, 40, 50]

# Min and Max
min_val = min(feature)
max_val = max(feature)

# Value to scale
x = 30

# Min-Max Scaling Formula
scaled_value = (x - min_val) / (max_val - min_val)

print("Min =", min_val)
print("Max =", max_val)
print("Scaled Value =", scaled_value)
```

**4. Standard Scaling**
**Definition**: Standard Scaling transforms data so that:

- Mean ≈ 0
- Standard Deviation ≈ 1

**Formula**

Formula
`X_scaled = X−μ / σ`
	​
where

- μ=Mean
- σ=Standard Deviation

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Create DataFrame
df = pd.DataFrame({
    'salary': [30000, 50000, 70000, 90000, 110000]
})

print("Original Data:")
print(df)

# Create StandardScaler Object
scaler = StandardScaler()

# Apply Standard Scaling
df['salary_scaled'] = scaler.fit_transform(
    df[['salary']]
)

print("\nAfter Standard Scaling:")
print(df)

# Verify Mean and Standard Deviation
print("\nMean:", round(df['salary_scaled'].mean(), 4))
print("Std :", round(df['salary_scaled'].std(), 4))
```

**5. Min-Max vs Standard Scaling**
| Feature | Min-Max Scaling   | Standard Scaling |
| ------- | ----------------- | ---------------- |
| Range   | [0,1]             | Not fixed        |
| Minimum | 0                 | Not fixed        |
| Maximum | 1                 | Not fixed        |
| Mean    | Not necessarily 0 | ≈ 0              |
| Std Dev | Not necessarily 1 | ≈ 1              |
| Formula | (x-min)/(max-min) | (x-mean)/std     |

# Topic 6: Creating New Features and Date-Time Feature Engineering

New features are created by combining or transforming existing columns;

**examples from the session:**

`average_order_value = total_revenue / total_orders ,`

`salary_per_year_of_experience = salary / years_of_experience` .

- `pd.cut()` converts a continuous numerical column into categorical bins with custom labels — for example, **grouping age** into `"teen"`, `"young"`, `"adult"`, `"senior"` based on defined boundaries.
- Date columns stored as **strings should be converted to datetime objects** using `pd.to_datetime()` ; from a **datetime column**, **features** like `year`, `month`, `day`, and `day-of-week` can be extracted using '.dt.year` , `.dt.month` , `.dt.day` , `.dt.dayofweek` .

**1. Creating New Features**

## How ?

**Step 1:** Convert String to Datetime.
pd.to_datetime()

**Example**

Now Pandas understands the **column as a date**.

After conversion: `df['order_date']` contains datetime objects.

```python
df['year'] = df['join_date'].dt.year
print(df['year'] )

df['month'] = df['join_date'].dt.month
print(df['month'])

df['day'] = df['join_date'].dt.day
print(df['day'])
```

**Extract Year**

`df['day_of_week'] = df['order_date'].dt.dayofweek`

```python
df['day_of_week'] = (
    df['join_date']
    .dt.dayofweek
)
print(df['day_of_week'])
```

**Why is This Useful?**

Business questions such as: Which day generates maximum revenue?

**cannot be answered directly** from a raw date string.

```python
# Mini Project 1: Restaurant Customer Behavior Analysis (Complete Solution in One Cell)

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Q1: Display first 5 rows
print("First 5 Rows:")
print(tips.head())

# Q2: Scatter Plot (total_bill vs tip)
plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Q2: Total Bill vs Tip")
plt.show()

# Q3: Scatter Plot with hue='sex'
plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
plt.title("Q3: Total Bill vs Tip by Gender")
plt.show()

# Q4: Average Total Bill by Day
plt.figure(figsize=(6,4))
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Q4: Average Total Bill by Day")
plt.show()

# Q5: Total Revenue by Day
plt.figure(figsize=(6,4))
sns.barplot(x='day', y='total_bill', estimator=sum, data=tips)
plt.title("Q5: Total Revenue by Day")
plt.show()

print("\nRevenue by Day:")
print(tips.groupby('day')['total_bill'].sum())

# Q6: Count Plot
plt.figure(figsize=(6,4))
sns.countplot(x='day', data=tips)
plt.title("Q6: Customer Count by Day")
plt.show()

print("\nCustomer Count by Day:")
print(tips['day'].value_counts())

# Q7: Histogram
plt.figure(figsize=(6,4))
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title("Q7: Distribution of Total Bill")
plt.show()

# Q8: Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(y='total_bill', data=tips)
plt.title("Q8: Box Plot of Total Bill")
plt.show()

# Q9: Grouped Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x='sex', y='tip', data=tips)
plt.title("Q9: Tip Distribution by Gender")
plt.show()

# Q10: Correlation Heatmap
corr_matrix = tips.corr(numeric_only=True)

print("\nCorrelation Matrix:")
print(corr_matrix)

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='Blues')
plt.title("Q10: Correlation Heatmap")
plt.show()

# BONUS Q1: Highest Revenue Day
revenue = tips.groupby('day')['total_bill'].sum()
print("\nRevenue by Day:")
print(revenue)
print("\nHighest Revenue Day:", revenue.idxmax())

# BONUS Q2: Higher Average Tip by Gender
avg_tip = tips.groupby('sex')['tip'].mean()
print("\nAverage Tip by Gender:")
print(avg_tip)
print("\nHigher Average Tip Given By:", avg_tip.idxmax())

# BONUS Q3: Weekend vs Weekday Revenue
tips['is_weekend'] = tips['day'].isin(['Sat', 'Sun'])

print("\nWeekend vs Weekday Revenue:")
print(tips.groupby('is_weekend')['total_bill'].sum())

plt.figure(figsize=(6,4))
sns.barplot(x='is_weekend', y='total_bill', estimator=sum, data=tips)
plt.title("BONUS: Weekend vs Weekday Revenue")
plt.show()

# BONUS Q4: Revenue by Smoker Category
plt.figure(figsize=(6,4))
sns.barplot(x='smoker', y='total_bill', estimator=sum, data=tips)
plt.title("BONUS: Revenue by Smoker Category")
plt.show()
```

**Mini Project 2:** Customer Feature Engineering and Scaling

**Dataset**: Create the following DataFrame:

```python
import pandas as pd

df = pd.DataFrame({
    'customer_id':[1,2,3,4,5],
    'age':[22,35,41,28,55],
    'annual_income':[300000,800000,1200000,600000,1500000],
    'total_orders':[10,25,40,18,50],
    'total_revenue':[50000,250000,600000,150000,900000],
    'join_date':['2022-01-15',
                 '2021-07-20',
                 '2020-03-10',
                 '2023-05-12',
                 '2019-09-18']
})
```

**Tasks**
- **Q1.** Create a new feature:

- average_order_value = total_revenue / total_orders

- **Q2.** Convert join_date into datetime format.

- **Q3.** Extract: year,
month,
day,
day_of_week

from join_date.

- **Q4.** Create a new feature: is_weekend using day_of_week.

- **Q5.** Convert age into categories using pd.cut():

- 0-25     → Young
- 25-40    → Adult
- 40-60    → Senior

Store in: age_group

- **Q6.** Apply Min-Max Scaling on: annual_income

- **Q7.** Apply Standard Scaling on: annual_income and verify:

mean ≈ 0,
std ≈ 1

- **Q8.** Create a new feature: log_revenue using: np.log()


- **Q9.** Generate a pair plot for all numerical variables.

- **Q10.** Generate a correlation heatmap.

```python
# Mini Project 2: Customer Feature Engineering and Scaling (Complete Solution in One Cell)

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler, StandardScaler

# --------------------------------------------------
# Create Dataset
# --------------------------------------------------
df = pd.DataFrame({
    'customer_id':[1,2,3,4,5],
    'age':[22,35,41,28,55],
    'annual_income':[300000,800000,1200000,600000,1500000],
    'total_orders':[10,25,40,18,50],
    'total_revenue':[50000,250000,600000,150000,900000],
    'join_date':['2022-01-15',
                 '2021-07-20',
                 '2020-03-10',
                 '2023-05-12',
                 '2019-09-18']
})

print("Original Dataset")
print(df)

# --------------------------------------------------
# Q1: Create Average Order Value
# --------------------------------------------------
df['average_order_value'] = (
    df['total_revenue'] /
    df['total_orders']
)

# --------------------------------------------------
# Q2: Convert Join Date to Datetime
# --------------------------------------------------
df['join_date'] = pd.to_datetime(df['join_date'])

# --------------------------------------------------
# Q3: Extract Date Features
# --------------------------------------------------
df['year'] = df['join_date'].dt.year
df['month'] = df['join_date'].dt.month
df['day'] = df['join_date'].dt.day
df['day_of_week'] = df['join_date'].dt.dayofweek

# --------------------------------------------------
# Q4: Create is_weekend Feature
# --------------------------------------------------
df['is_weekend'] = np.where(
    df['day_of_week'] >= 5,
    1,
    0
)

# --------------------------------------------------
# Q5: Create Age Groups using pd.cut()
# --------------------------------------------------
df['age_group'] = pd.cut(
    df['age'],
    bins=[0,25,40,60],
    labels=['Young','Adult','Senior']
)

# --------------------------------------------------
# Q6: Min-Max Scaling
# --------------------------------------------------
minmax_scaler = MinMaxScaler()

df['income_minmax'] = minmax_scaler.fit_transform(
    df[['annual_income']]
)

# --------------------------------------------------
# Q7: Standard Scaling
# --------------------------------------------------
standard_scaler = StandardScaler()

df['income_standard'] = standard_scaler.fit_transform(
    df[['annual_income']]
)

print("\nMean After Standard Scaling:")
print(round(df['income_standard'].mean(),4))

print("\nStd After Standard Scaling:")
print(round(df['income_standard'].std(),4))

# --------------------------------------------------
# Q8: Log Transformation
# --------------------------------------------------
df['log_revenue'] = np.log(
    df['total_revenue']
)

# --------------------------------------------------
# Display Updated Dataset
# --------------------------------------------------
print("\nUpdated Dataset")
print(df)

# --------------------------------------------------
# Q9: Pair Plot
# --------------------------------------------------
numerical_cols = [
    'age',
    'annual_income',
    'total_orders',
    'total_revenue',
    'average_order_value'
]

sns.pairplot(df[numerical_cols])
plt.show()

# --------------------------------------------------
# Q10: Correlation Heatmap
# --------------------------------------------------
corr_matrix = df.corr(numeric_only=True)

print("\nCorrelation Matrix")
print(corr_matrix)

plt.figure(figsize=(10,7))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='Blues'
)

plt.title("Customer Feature Engineering Heatmap")
plt.show()

# --------------------------------------------------
# BONUS ANALYSIS
# --------------------------------------------------

print("\nAverage Revenue by Age Group")
print(
    df.groupby('age_group')['total_revenue']
      .mean()
)

print("\nWeekend Customers")
print(
    df[df['is_weekend']==1]
)

print("\nHighest Revenue Customer")
print(
    df.loc[df['total_revenue'].idxmax()]
)
```
