**1.1 What is a Dictionary?**

A dictionary stores data in key-value pairs.

```python
#Syntax
variable_name = {"key1":"Value",
                 "key2":"value"}


#Example
student = {
    "name": "Rahul",
    "marks": 90
}
print(student)

#"name" → key
#"Rahul" → value

# Important Points need to keep
# Uses {} curly brackets
# Keys should be unique
# Values can be any data type
# Dictionaries are mutable (can change)
```

**Accessing Values**

```python
print(student["name"])
#print(student["age"])

    # OR
#dictionary.get(key, default_value)   # syntax
print(student.get("age", "Not Found"))
```

**Adding New Data**

```python
student["city"] = "Delhi"
print(student, " - Student Dictionary")
```

**Deleting Data**

```python
print(student,"Before deleting marks data from dictionary")
del student["marks"]
print(student, "After deleting marks data from dictionary")
```

**Nested Dictionary**

A nested dictionary means a **dictionary inside another dictionary**.

It is used to **store** detailed or **grouped information** in an organized way.


```python
students = {
    "student1": {
        "name": "Rahul"
    }
}

# Access the name Rahul
print(students["student1"]["name"]) #[] brackets to access values using keys in a dictionary.
```

# **2. Set Operations**

- Set operations are operations used to compare or combine sets.

**Main set operations:**

- Union (|) → combines all unique elements
- Intersection (&) → common elements
- Difference (-) → elements present in one set only

**Union** (|) → combines all unique elements

```python
a = {1, 2, 2, 3}
b = {3, 4, 5, 6}
union = a | b
print("a",a)
print("b",b)
print("Union=", union)
```

**Intersection** (&) → common elements

```python
intersection = a & b
print("a",a)
print("b",b)
print("Intersection(&) gives common element = ", intersection)
```

**Difference** (-) → elements present in one set only

```python
difference1 = a - b #1 and 2 are only in a
difference2 = b - a #1 and 2 are only in a

print("a",a)
print("b",b)
print("Difference = ", difference1)
print("Difference = ", difference2)
```

**What is Control Flow?**

Control flow is the **process of making decisions in a program** using *conditions* like **if, elif, and else**.

It helps the program make decisions.

**if Statement** - Runs only if condition is True.

```python
age = 20

if age > 18:
    print("Adult")
```

**elif statement** - Checks another condition if previous one is False.


```python
light = "yellow"

if light == "green":
    print("Go")

elif light == "yellow":
    print("Slow")
```

**else Statement** - Runs when all above conditions are False.

```python
if age > 18:
    print("Adult")

else:
    print("Minor")
```

**Indentation** - Python uses spaces/tabs to define blocks.

Wrong indentation gives error.

```python
if  age>18:
print("Adult")
```

```python
if  age>18:
  print("Adult")
```

**1.4 Logical and Membership Operators**

- **and Operator** : Both conditions must be True.
- **or Operator** : At least one condition should be True.
- **not Operator** : Reverses the condition.
- **in Operator** - Checks whether item exists in collection.

**and Operator :** Both conditions must be True.

# 4. **Subjective Question : E-commerce Checkout Logic**

**Scenario:** An e-commerce platform needs a new checkout module. The system determines the final order total by applying sequential shipping rules and loyalty discounts based on the order amount, customer loyalty tier, and
destination (International vs. Domestic).

**International Shipping:** Base is $25.00. If order > $150.00, shipping is $10.00. VIP customers always get $0.00 shipping.


**Domestic Shipping:** Base is $10.00. If order >= $50.00, shipping is $0.00.

Premium and VIP customers always get $0.00.


**Discounts:** (Applied to the subtotal of Order Amount + Shipping). "Standard" gets 0\%. "Premium" gets 5\% ONLY if the subtotal is > $100.00. "VIP" gets 10\%, but if their base order (before shipping) was > $200.00, they get 15\%
instead.

**Task: Write nested conditional logic using if , elif , and else to calculate the final total. If an unrecognized tier is provided (not Standard, Premium, or VIP), return -1.0 . The final amount must be rounded to 2 decimal places.**

**Step-by-step approach:**

1. **First, validate the input tier**. Check if it's in a valid set or list (Standard, Premium, VIP). If not, immediately
return -1.0 .
2. **Evaluate the shipping logic.** Use an if/else block to split International vs. Domestic. Inside each, **use if/elif/else** to assign the proper shipping cost based on tier and base order amount.
3. **Calculate the subtotal** by adding the base order amount and the calculated shipping cost.
4. **Evaluate the discount logic.** Use if/elif based on the tier. Apply the conditional logic checking the subtotal
or base order amount thresholds as specified.
5. **Multiply the subtotal** by the appropriate discount factor, round to 2 decimal places, and return the result

**Soluton 1**

```python
order_amount = 120 # Stores the order price in Order_amount variable.
customer_tier = "Premium"  # Stores customer type.
is_international = True  # Checks whether order is international or domestic.

# Step 1: Check valid customer
if customer_tier not in ["Standard", "Premium", "VIP"]: #Checks whether customer tier is valid.
    print(-1.0) # Prints error value if tier is invalid.

else:

    # Step 2: Shipping Costs
    if is_international: # Checks if shipping is international.

        if customer_tier == "VIP": # VIP international customers get free shipping.
            shipping = 0

        elif order_amount > 150: # If international order is above 150, shipping becomes 10.
            shipping = 10

        else:
            shipping = 25  # Otherwise normal international shipping is 25.

    else:

        if customer_tier == "Premium" or customer_tier == "VIP": #Premium and VIP domestic customers get free shipping.
            shipping = 0

        elif order_amount >= 50: # Domestic orders above 50 get free shipping.
            shipping = 0

        else:
            shipping = 10 # Otherwise domestic shipping is 10.

    # Step 3: Subtotal
    subtotal = order_amount + shipping # Adds order amount and shipping cost.

    # Step 4: Discount
    discount = 0 # Initial discount is 0%.

    if customer_tier == "Premium" and subtotal > 100: # Premium customers get 5% discount if subtotal is above 100.
        discount = 0.05

    elif customer_tier == "VIP":  # Checks if customer is VIP.

        if order_amount > 200: # VIP customers get 15% discount above 200 order amount.
            discount = 0.15

        else:
            discount = 0.10 # Otherwise VIP gets 10% discount.

    # Step 5: Final Amount
    final_amount = subtotal * (1 - discount) # Calculates final amount after discount.

    print(round(final_amount, 2)) # Prints final answer rounded to 2 decimal places.
```

**Solution 2**

```python
def process_order(order_amount, customer_tier, is_international):

    # Step 1: Validate Tier
    valid_tiers = {"Standard", "Premium", "VIP"}

    if customer_tier not in valid_tiers:
        return -1.0

    # Step 2: Calculate Shipping
    shipping_cost = 0.0

    if is_international:

        if customer_tier == "VIP":
            shipping_cost = 0.0

        elif order_amount > 150.0:
            shipping_cost = 10.0

        else:
            shipping_cost = 25.0

    else:

        if customer_tier == "VIP" or customer_tier == "Premium":
            shipping_cost = 0.0

        elif order_amount >= 50.0:
            shipping_cost = 0.0

        else:
            shipping_cost = 10.0

    # Step 3: Calculate Subtotal
    subtotal = order_amount + shipping_cost

    # Step 4: Apply Discount Logic
    discount = 0.0

    if customer_tier == "Premium":

        if subtotal > 100.0:
            discount = 0.05

    elif customer_tier == "VIP":

        if order_amount > 200.0:
            discount = 0.15

        else:
            discount = 0.10

    # Step 5: Final calculation
    final_amount = subtotal * (1 - discount)

    return round(final_amount, 2)
```

```python
# Test cases:
test1 = process_order(120.0, "Premium", True) # Expected: 137.75
test2 = process_order(250.0, "VIP", False) # Expected: 212.5
test3 = process_order(45.0, "Guest", False) # Expected: -1.0
print(test1,"test1")
print(test2,"test2")
print(test3,"test3")
```
