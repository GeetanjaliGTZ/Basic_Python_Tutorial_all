## 1. For Loops and Iteration, The range() Function

**What is a Loop?**

A loop is used to repeat the same task multiple times.

```python
for i in range(4):
    print("Hello")

```

**What is Iteration?**

Each single run of the loop is called an iteration.

```python
for i in range(3):
    print(i)

# 1st iteration → i = 0
# 2nd iteration → i = 1
# 3rd iteration → i = 2
```

**Syntax of for Loop**

```python
for variable in iterable:
    code
```

```python
for num in [1, 2, 3]:
    print(num)
```

**Using range()**

range() generates numbers.

```python
for i in range(5):
    print(i)
```

**range(start, stop, step)**

```python
for i in range(1, 10, 2):
    print(i)

# add 2 each time
# 1 + 2 = 3
# 3 + 2 = 5
# 5 + 2 = 7
```

**Reverse Loop**

Using negative step:

```python
for i in range(5, 0, -1):
    print(i)
```

**Looping Through List**

```python
names = ["A", "B", "C"]

for name in names:
    print(name)
```

**Looping Through Dictionary**

```python
student = {
    "name": "Ritesh",
    "score": 90
}

for key, value in student.items():
    print(key, value)
```

**enumerate()**

Used to get: index, value

```python
names = ["A", "B", "C"]

for index, name in enumerate(names):
    print(index, name)
```

**break**

Stops the loop completely.

```python
for i in range(5):
    if i == 3:
        break
    print(i)
```

**continue**

Skips current iteration.

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

**Nested Loop**

Loop inside another loop.

```python
for i in range(2):
    for j in range(3):
        print(i, j)
```

```python
classes = ["math", "English"]

Students = [
    ["Rahul", "priya", "Geetanjali"],
    ["Neha", "Rohit", "Neetu"]
]
print("len of classes", len(classes) )


for i in range(len(classes)):
  print(classes[i])
  for student in Students[i]:
    print("Students:", student)
```

```python
for i in range(len(classes)):
  print(classes[i])
```

```python
for student in Students[i]:
  print("Students:", student)
```

- **Loop** = repeating work
- **Iteration** = one repetition
- **break** = stop completely
- **continue** = skip current step
- **enumerate()** = position + value together

## 2. Iterating Over Dictionaries

**What is a Dictionary?**

A dictionary stores data in: key : value pairs

```python
student = {
    "name": "Ritesh",
    "score": 90
}
```

**What Does "Iterating Over Dictionary" Mean?**

It means looping through dictionary data one by one.

**Iterating Over Keys**

```python
student = {
    "name": "Ritesh",
    "score": 90
}

for key in student:
    print(key)
# By default, loop gives keys.
```

**Iterating Over Values**

Using .values()

```python
for value in student.values():
    print(value)
```

**Iterating Over Both Key and Value**
Using .items()

```python
for key, value in student.items():
    print(key, value)
```

**Example**

```python
employee = {
    "name": "Aman",
    "salary": 50000
}

for key, value in employee.items():
    print(f"{key} -> {value}")
```

- **.keys()** → only keys
- **.values()** → only values
- **.items()** → both key and value

## 3. While Loops and Control Statements

**What is a while Loop?**

A while loop runs repeatedly until a condition becomes False.

**Syntax**


```python
while condition:
    code
```

**Basic While Loop**

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

**Explanation**
- loop starts from 1
- runs while condition is True
- count += 1 increases value each iteration

**Infinite Loop**

If condition never becomes False, loop runs forever.

```python
count = 1

while count > 0:
    print(count)
```

This is called an infinite loop ❌

**Control Statements**

Control statements change loop behavior.

**Main control statements:**

 - break
 - continue
 - pass

**break**

Stops the loop completely.

```python
for i in range(5):
    if i == 3:
        break
    print(i)
```

**continue**

Skips current iteration.

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
# Here 2 is skipped.
```

**pass**

pass means “do nothing”.

```python
for i in range(3):
    pass
```

Used as placeholder when code is incomplete.

- **while** → repeat until condition fails
- **break** → stop completely
- **continue** → skip one step
- **pass** → do nothing temporarily

## 4. Functional Programming (Functions and Returns)

**What is a Function?**

A function is a **reusable block of code** used to perform a task.

Instead of *writing the same code again and again, we create a function*.

**Function Definition**

We use **def** to create a function.

**Syntax**

```python
def function_name():
    code
```

**Example**

```python
def greet():
    print("Hello")
```

- **def** → define function
- **greet** → function name

**Function Calling**

To run the function:

```python
greet()
```

**Function with Parameters**

Parameters are inputs given to a function.

```python
def add(a, b):
    print(a + b)

add(2, 3)

# a and b are parameters
# 2 and 3 are arguments
```

**Return Statement**

return sends a value back from the function.

```python
def add(a, b):
    return a + b

result = add(2, 3)

print(result)
```

**Difference Between print() and return()**

**Using print()**

Displays output directly.

```python
def add(a, b):
    print(a + b)
```

**Using return**

Returns value so it can be stored or reused.

```python
def add(a, b):
    return a + b
```

**Example**

```python
def average_salary(salaries):
    return sum(salaries) / len(salaries)

masai = [100000, 200000, 40000]

avg = average_salary(masai)

print(avg)
```

- **def** → create function
- **function call** → run function
- **parameters** → inputs
- **return** → send result back

# Solution Problem 1

```python
orders = [
    {"customer_id": "C001", "order_amount": 1200, "status": "OK"},
    {"customer_id": "C002", "order_amount": 0, "status": "OK"},
    {"customer_id": "C003", "order_amount": 3400, "status": "OK"},
    {"customer_id": "C004", "order_amount": 500, "status": "OK"},
    {"customer_id": "C005", "order_amount": 800, "status": "FRAUD"},
    {"customer_id": "C006", "order_amount": 2100, "status": "OK"},
]

total = 0

for order in orders:

    # Skip invalid orders
    if order["order_amount"] == 0:
        continue

    # Stop processing if fraud found
    if order["status"] == "FRAUD":
        break

    # Add valid order amount
    total += order["order_amount"]

print("Total valid orders processed:", total)
```

## Solution Problem 2


```python
# Constant
MONTHLY_TAX_THRESHOLD = 600000 / 12

# Employee Data
employees = [
    ("Alice", 80000),
    ("Bob", 200000),
    ("Carol", 45000),
    ("Dave", 120000),
    ("Eve", 30000)
]

# Salary List
salaries = [80000, 200000, 45000, 120000, 30000]


# Function 1: Count Taxed Employees
def count_taxed(employees):
    count = 0

    for name, salary in employees:
        if salary > MONTHLY_TAX_THRESHOLD:
            count += 1

    return count


# Function 2: Calculate Total Salary
def total_salary(salaries):
    total = 0

    for salary in salaries:
        total += salary

    return total


# Function 3: Calculate Average Salary
def average_salary(salaries):
    total = total_salary(salaries)
    average = total // len(salaries)

    return average


# Function Calls
taxed = count_taxed(employees)
total = total_salary(salaries)
average = average_salary(salaries)


# Final Output
print("Taxed employees:", taxed)
print("Total salary:", total)
print("Average salary:", average)
```
