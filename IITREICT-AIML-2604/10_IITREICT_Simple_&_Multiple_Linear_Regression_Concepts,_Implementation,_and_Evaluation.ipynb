# Topic 1: Regression vs Classification

**1. Real-Life Example**

Imagine you are working for an HR company.

**Scenario 1**

Your manager asks: "Predict the salary of a candidate with 5 years of experience."

Possible outputs:
₹45,000
₹52,000
₹61,750
₹70,500

**Notice**: The output can be any numerical value. This is a **Regression Problem**.

**Scenario 2**

Now your manager asks: "Will the candidate be selected?"

Possible outputs: Selected, Not Selected.

Only fixed categories exist. This is a **Classification Problem.**

**2. Key Rule**

The distinction depends entirely on the **Target Variable (Y)**. Not on the input features.

As stated in the tutorial, the decision is based on the **nature of Y**, **not X**.

**3. What is Regression?**

**Definition** : Regression is used when the target variable is continuous.

**Continuous means:** Can take any value within a range.

**Example**: -

| Problem                | Output     |
| ---------------------- | ---------- |
| Salary Prediction      | ₹50,000    |
| House Price Prediction | ₹75,00,000 |
| Temperature Prediction | 34.7°C     |
| Solar Energy Output    | 225.6 kWh  |



**Example**

**Input**: Experience = 5 Years

**Output**: Salary = ₹60,000

Since salary can take any value: **Regression**

**4. What is Classification?**

**Definition**: Classification is used when the **target** variable belongs to a **fixed set of categories**.

**Examples**:

| Problem            | Output                        |
| ------------------ | ----------------------------- |
| Spam Detection     | Spam / Not Spam               |
| Tumor Detection    | Benign / Malignant            |
| Sentiment Analysis | Positive / Negative / Neutral |
| Student Result     | Pass / Fail                   |


**5. Continuous vs Discrete Outputs**

**Regression** (can take any value within a range, including decimal values - **Continuous**) : 25.5, 30.7, 45.2, 67.8. Infinite possible values.

**Classification** (fixed number of categories or classes - **discrete/categorical** ) : Pass, Fail or Positive, Negative, Neutral
Limited categories.

**Examples:**

| Output Variable (Y)           | Continuous or Discrete? | Problem Type   |
| ----------------------------- | ----------------------- | -------------- |
| Salary = ₹60,000              | Continuous              | Regression     |
| House Price = ₹50 Lakh        | Continuous              | Regression     |
| Temperature = 35.8°C          | Continuous              | Regression     |
| Pass / Fail                   | Discrete                | Classification |
| Spam / Not Spam               | Discrete                | Classification |
| Benign / Malignant            | Discrete                | Classification |
| Positive / Negative / Neutral | Discrete                | Classification |


# Topic 2: Supervised vs Unsupervised Learning

**1. Real-Life**

Imagine you have employee records:

| Experience | Education | Salary |
| ---------- | --------- | ------ |
| 2          | B.Tech    | 35000  |
| 5          | M.Tech    | 65000  |
| 8          | PhD       | 100000 |

Notice:

**Inputs (X)**  → Experience, Education

**Output (Y)**  → Salary

Since we know both inputs and outputs, the model can learn directly. This is **Supervised Learning**

**Now imagine a hospital dataset:**

| Age | Blood Pressure | Weight |
| --- | -------------- | ------ |
| 35  | 120            | 65     |
| 50  | 140            | 82     |
| 60  | 160            | 90     |

But we do NOT know: Disease, Healthy, Cancer, Diabetes
**No labels are available.** This becomes **Unsupervised Learning**

**2. What is Supervised Learning?**

**Definition**: Supervised Learning uses:
X (Features) + Y (Target) during training.

The model learns: Input → Output relationship.

The tutorial defines supervised learning as **learning from labeled data** where both **X** and **Y** are **available**.

**3. What is Unsupervised Learning?**

**Definition**: Unsupervised Learning uses only: X but **No target variable is available.**

The model tries to discover patterns on its own.

The tutorial describes this as a situation where **Y labels are missing**.

**4. Labeled vs Unlabeled Data**

**Labeled Dataset Contains:** Features (X) + Target (Y)

**Example**:

| Study Hours | Pass |
| ----------- | ---- |
| 2           | 0    |
| 5           | 1    |
| 8           | 1    |

Here: *Pass is the label*.

**Therefore**: Labeled Dataset

**Unlabeled Dataset Contains:** Features (X) only.

**Example**:

| Study Hours |
| ----------- |
| 2           |
| 5           |
| 8           |

*No target column exist*s.

**Therefore**: Unlabeled Dataset

**7. Python Example: Supervised Learning**

```python
from sklearn.linear_model import LinearRegression

# Features
X = [[1], [2], [3], [4]]

# Labels
y = [30000, 40000, 50000, 60000]

model = LinearRegression()

model.fit(X, y)

print(model.predict([[5]]))
```

**8. Python Example: Unsupervised Learning**

```python
from sklearn.cluster import KMeans
import pandas as pd

data = pd.DataFrame({
    'Age':[22,25,50,55],
    'Income':[25000,30000,90000,95000]
})

model = KMeans(n_clusters=2)

model.fit(data)

print(model.labels_)

# Notice: No Y variable exists. The algorithm created groups automatically.
```

**9. Which One Should We Prefer?**

If both **X and Y are available**, prefer **supervised learning** because the model has more information and generally achieves better accuracy.

**Example:**

| Problem                    | Supervised / Unsupervised |
| -------------------------- | ------------------------- |
| Salary Prediction          | Supervised                |
| House Price Prediction     | Supervised                |
| Spam Detection             | Supervised                |
| Customer Segmentation      | Unsupervised              |
| Grouping Similar Customers | Unsupervised              |
| Clustering Documents       | Unsupervised              |


**Summary**: Supervised Learning uses labeled data (X + Y) to learn input-output relationships, whereas Unsupervised Learning uses only X and discovers patterns without labels. Linear Regression is a supervised learning algorithm because it requires both features and target values during training.

# Topic 3: Simple Linear Regression

**1. Real-Life Example**

Suppose an HR company has the following employee records:

| Experience (Years) | Salary  |
| ------------------ | ------- |
| 1                  | ₹28,000 |
| 2                  | ₹36,000 |
| 3                  | ₹44,000 |
| 4                  | ₹52,000 |
| 5                  | ₹60,000 |

Now a **new candidate arrives** with: *Experience = 6 years*

The **question is**: **What salary should we offer ?**

Instead of guessing, we want a mathematical formula.

That **formula is learned by**: *Simple Linear Regression*

**2. What is Simple Linear Regression?**

**Definition**: Simple Linear Regression models the relationship between: One Input Feature (X) -> One Continuous Output (Y)

This **defines it as** a relationship between one input feature and a continuous output variable.

**3. Linear Regression Equation**

The most important equation:

y^ = w0 + w1 x

or

y^ = b + mx

where:
| Symbol    | Meaning             |
| --------- | ------------------- |
| (y^)      | Predicted Output    |
| (w_0)     | Intercept           |
| (w_1)     | Coefficient (Slope) |
| (x)       | Input Feature       |


**4. Understanding the Intercept (w0)**

**Intercept means:** Predicted value when X = 0

**Example**:

Salary=20000+8000×Experience

When: Experience = 0, Salary=20000

Therefore: Intercept = 20,000

**Intercept = Starting Value**

**5. Understanding the Coefficient (w1)**

**Coefficient tells us:** How much Y changes when X increases by 1 unit

**Example**:

Salary=20000 + 8000 × Experience

Coefficient: 8000

**Meaning**: Every extra year of experience increases salary by:
₹8,000

**Coefficient = Rate of Change**

**7. Goal of the Algorithm**

The algorithm tries to find the optimal values of: w₀, w₁
such that prediction errors are minimized.

In simple words: **Find the Best Line that fits the data.**

**8. Visual Understanding**

Salary
 ^
 |
 |                  *
 |              *
 |          *
 |      *
 |   *
 +----------------------> Experience

 Linear Regression tries to **draw the best straight line** through the points.

**9. Python Example**

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# create the dataset
data = {
    'Experience':[1,2,3,4,5],
    'Salary':[28000,36000,44000,52000,60000]
}
df = pd.DataFrame(data)

# Step 3: Create X and y
X = df[['Experience']]
y = df['Salary']

# Step 4: Create Model
model = LinearRegression()

# Step 5: Train Model
model.fit(X, y)   # fit() learn from Data

# Step 6: View Coefficient and Intercept
print("Coefficient:", model.coef_[0])  # Every additional year of experience increases salary by ₹8,000.
print("Intercept:", model.intercept_)  # Base salary when experience = 0.

prediction = model.predict([[6]]) # Make Predictions, Predict salary for: Experience = 6
# Use Learned Knowledge to make new predictions.

print("Predicted Salary:", prediction[0])  #
```
