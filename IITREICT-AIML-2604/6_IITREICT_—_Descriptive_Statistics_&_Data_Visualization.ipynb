**Topic 1: Central Tendency**
Definition: Statistical values used to find the center or typical value of a dataset.

- **Mean & Median:** The mean is the **average**, calculated by *summing all values and dividing by the count*. The
median is the exact middle value of a sorted dataset.
- **Use-Case:** The median is preferred for representing typical worker salaries or delivery times because it is not affected by extreme outliers, whereas the mean is highly sensitive to them.
- **Mode:** The **most frequent value**, which is particularly useful for finding the most common categorical data (like the most used payment method).

**Why Are We Learning This?**

Imagine you are working at:

 - Swiggy/Zomato → Need typical delivery time
 - Amazon → Need average order value
 - School → Need average marks

If there are **thousands of values**, **looking at every value is impossible**.

We need **one number** that **represents** the **dataset**.

That is why we use Central Tendency.

**What is Central Tendency?**

- **Definition:** Central Tendency tells us the **center** or typical value of a dataset.

```python
# Example
marks = [70, 75, 80, 85, 90] # A single number should represent these marks.
```

**Types of Central Tendency**

- There are **3 measures**: Mean, Median, Mode

**1. Mean (Average)**

***Formula***

- **Mean**=  Number of values / Sum of all values
	​


```python
#Example
marks = [70, 80, 90]

mean = (70 + 80 + 90) / 3

print(mean)
# Simple Explanation (Mean = Add everything ÷ Total count)
```

```python
# Real life Example (Monthly sales:)
sales = [100, 120, 130]
mean = (100+120+130)/3
print("mean", mean)
```

**2. Median**

**Topic 2: Measures of Spread**

- **Definition**: Metrics that measure how much the data varies or spreads out.
Range: The simplest measure, calculated as the maximum value minus the minimum value.
- **Variance & Standard Deviation:** Variance measures the average squared deviation from the mean. Standard
deviation is simply the square root of the variance.
- **Nuance:** We use standard deviation because taking the square root brings the spread measurement back into the
original units of the data (e.g., runs or centimeters instead of squared runs).

**Why Are We Learning This?**

Imagine two batsmen:

| Player | Average Runs |
| ------ | ------------ |
| A      | 50           |
| B      | 50           |

Both have the same average.

**Can we say both play the same?**

❌ No

**Let's look at their scores:**



```python
Player_A = [48, 50, 52, 49, 51]
Player_B = [5, 100, 20, 80, 45]
```

Both have **average** ≈ 50.

**But:**

- Player A is very **consistent**.
- Player B is **highly inconsistent**.

We need a **way to measure how spread out** the data is.

This is **called Measures of Spread.**

**What is Spread?**

- Definition : Spread tells us **how much the data varies from its center**.

```python
# Example
dataset_1 = [49, 50, 51]  # Very close together, Small spread.
dataset_2 = [10, 50, 90]  # Far apart, Large spread.
print("dataset_1", dataset_1)
print("dataset_1", dataset_2)
```

**Types of Measures of Spread:**

- Range
- Variance
- Standard Deviation

**1. Range**

- **Definition**: Range is the difference between: **Maximum Value - Minimum Value**
- **Formula:** Range = Max − Min

```python
#Example
marks = [50,60,70,80,90]  # Maximum: 90, MInimum: 50
Range = 90 - 50
print("Range", Range)
```

```python
marks = [50,60,70,80,90]

print("Range", max(marks)-min(marks))
```

**Limitation of Range**

**Range only looks at:**

- smallest value
- largest value

**It ignores everything else.**

**2. Variance**

- **Definition** : Variance measures **how far value**s are **from the mean**.

- **Variance tells us:** "How much do values move away from the average?"

**Example**
data = [10,20,30]

- Step 1: Find Mean
      Mean = 20

- Step 2: Difference from Mean
      10 - 20 = -10

      20 - 20 = 0

      30 - 20 = 10

- Step 3: Square Differences
      (-10)^2 = 100

      0^2 = 0

      10^2 = 100

- Step 4: Average of Squared Differences
(100 + 0 + 100)/3  = 66.67 (Variance)


**3. Standard Deviation**

- **Definition:** Standard deviation is the **square root of variance**.

**Formula**

Standard Deviation= **square root of Variance**
	​


**Example**

If
Variance = 64
then
Standard Deviation = 8

- Standard deviation tells us whether the values **stay close to the average** or are **spread far away** from it.

**Simple Interpretation**

- **Small Standard Deviation :** Data points close to mean (Consistent data).
- **Large Standard Deviation :** Data points far from mean (Inconsistent data).

**Cricket Example**
- **Player A:** Average = 50, Standard Deviation = 5
- **Player B:** Average = 50, Standard Deviation = 40

**Question: Who is more consistent?**

**Answer :**
Player A, Because lower standard deviation means less variation.

**Comparison**

| Metric             | Meaning                   |
| ------------------ | ------------------------- |
| Range              | Max - Min                 |
| Variance           | Average squared deviation |
| Standard Deviation | Spread in original units  |


**Topic 3: Percentiles, Quartiles & Outlier Detection**

- **Definition:** Percentiles give a relative ranking (e.g., the 90th percentile means 90% of the data is below that
value). Quartiles divide data into four equal parts (25%, 50%, 75%).
- **IQR (Interquartile Range):** Calculated as Q3 (75th percentile) minus Q1 (25th percentile). It represents the
middle 50% spread of the data.
- **Outlier Formula:** An outlier is mathematically defined as any point below the lower limit ( Q1 - 1.5 *
IQR ) or above the upper limit ( Q3 + 1.5 * IQR ).
- **Connection:** The pandas.DataFrame.describe() function instantly gives you count, mean, standard
deviation, min, percentiles, and max for numerical columns.

**Why Are We Learning This?**

Imagine a student says: "I am in the 90th percentile."

**What does that mean?**

Does it mean:

90 marks? ❌
90 questions correct? ❌

No.

It **tells** us **how the student performed compared to others**.

This is where **Percentiles** and **Quartiles** help.

**1. Percentiles**

**Definition :**  Percentiles tell us the **relative position of a value** in a dataset.

**Example:** Suppose 100 students wrote an exam.

If Rahul is in the: 90th Percentile.

**It means:**

- Rahul scored higher than 90% of students.

- Only 10% students scored better.

**Easy Example**

**If** you are in: 80th percentile

**Then:**

 - 80% students scored below you
- 20% students scored above you

**2. Quartiles**
**Definition:** Quartiles divide data into 4 equal parts.

0%      25%      50%      75%      100%
|--------|--------|--------|--------|
       Q1       Q2       Q3

- **Q1 (First Quartile):** 25% of data lies below it.
- **Q2 (Second Quartile):** 50% of data lies below it. Q2 is actually: **Median**
- **Q3 (Third Quartile):** 75% of data lies below it.

```python
quartiles = [10,20,30,40,50,60,70,80]
# Q1 : 25th Percentile
# Q2 : Median
# Q3 : 75th Percentile
```

**Why Quartiles?**

They help us **understand how data is distributed**.

**3. Interquartile Range (IQR)**

**Definition:** **IQR** measures the **spread of the middle 50% of data**.

**Formula**
- IQR = Q3 − Q1

### Example
**Q1** = 20, **Q3** = 60

then

**IQR** = 60 - 20
        = 40

**Simple Meaning:** IQR shows **how spread out the middle half of the data** is.

**Why Use IQR?**

- Because it **ignores extreme values** (**outliers**).

- It **focuses** on the **normal data**.

**4. Outlier Detection**

**What is an Outlier?**

- A value that is **much larger** or **much smaller** than the rest.

**How Do We Detect Outliers?**

- Using IQR.

**Formula**
- Lower Limit = Q1 - 1.5 × IQR
- Upper Limit = Q3 + 1.5 × IQR

**Rule**

**If value** is:

- **Below Lower Limit** → Outlier
- **Above Upper Limit** → Outlier

**Why 1.5?**

**Why multiply by 1.5?**

Answer:

- It is a **standard statistical rule** used worldwide **for detecting outliers**.

- No need to memorize the reason, **just remember the formula**.

**Real-Life Example**

- Delivery_time = 30,32,35,33,180
- Most deliveries: 30-35 minutes
- But 180 is extremely large.
- Therefore 180 = Outlier

**Anticipated Doubt Areas**

> **Why do we multiply by 1.5 in the IQR limits formula?** (It is a standard statistical industry formula; we use it to establish boundaries for what is considered an outlier).

> **Variance vs Standard Deviation:** (Standard deviation is preferred because it brings the unit of spread back to the original unit of the data, undoing the "squaring" done by variance).

> **Histograms vs Bar Plots:** (Bar plots are for comparing text categories/labels, while histograms group continuous numerical values into "bins").

> **Median line positioning in Box Plots:** (The median line isn't always in the exact center of the box; its position simply shows where 50% of the data falls, depending on the data's skew).
