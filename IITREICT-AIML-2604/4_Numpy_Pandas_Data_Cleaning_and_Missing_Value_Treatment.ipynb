```python
import pandas as pd
import numpy as np

# Sample raw employee payroll DataFrame
df = pd.DataFrame({
    'employee_id': [101, 102, 103, 104, 105],
    'name': ['  john', 'MARY ', None, 'alice', ' Bob '],
    'department': ['finance', 'HR', None, 'it', 'MARKETING'],
    'age': [25, np.nan, 35, np.nan, 40],
    'salary': ['₹50,000', '₹60,000', None, '₹70,500', '₹80,000'],
    'joining_date': ['12/01/2023', '15/02/2022', None,
                     '31/12/2021', 'invalid_date']
})

print("Original DataFrame:")
print(df)

# ----------------------------------------------------
# Step 1: Clean text columns
# ----------------------------------------------------

name_clean = (
    df['name']
    .fillna('unknown')
    .astype(str)
    .str.strip()
    .str.title()
)

department_clean = (
    df['department']
    .fillna('unknown')
    .astype(str)
    .str.strip()
    .str.title()
)

# ----------------------------------------------------
# Step 2: Clean age column
# ----------------------------------------------------

age_median = np.median(df['age'].dropna())

age_clean = (
    df['age']
    .fillna(age_median)
    .astype(int)
)

# ----------------------------------------------------
# Step 3: Clean salary column
# ----------------------------------------------------

salary_clean = (
    df['salary']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
)

salary_clean = pd.to_numeric(
    salary_clean,
    errors='coerce'
)

salary_median = np.median(
    salary_clean.dropna()
)

salary_clean = salary_clean.fillna(
    salary_median
)

# ----------------------------------------------------
# Step 4: Clean joining date column
# ----------------------------------------------------

date_clean = pd.to_datetime(
    df['joining_date'],
    errors='coerce',
    dayfirst=True
)

date_clean = date_clean.fillna(
    pd.to_datetime('1900-01-01')
)

# ----------------------------------------------------
# Step 5: Assemble cleaned DataFrame
# ----------------------------------------------------

clean_df = pd.DataFrame({
    'employee_id': df['employee_id'],
    'name': name_clean,
    'department': department_clean,
    'age': age_clean,
    'salary': salary_clean,
    'joining_date': date_clean
})

# ----------------------------------------------------
# Step 6: Validation
# ----------------------------------------------------

print("\nCleaned DataFrame:")
print(clean_df)

print("\nNull Values Check:")
print(clean_df.isnull().sum())

print("\nDataFrame Information:")
clean_df.info()
```
